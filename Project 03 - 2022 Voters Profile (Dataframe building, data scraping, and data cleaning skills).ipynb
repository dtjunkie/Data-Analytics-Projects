```python
# Importing Pandas
import pandas as pd
```

```python
# Reads the json file embedded from a news website
voters_profile = pd.read_json("https://data.gmanetwork.com/gno/microsites/eleksyon2022/voters_profile/provinces.gz")

# Transposes the indexes and columns
voters_profile = voters_profile.T

# Resets the index of the dataframe
voters_profile = voters_profile.reset_index()

# Rename the first column from 'index' to 'province'
voters_profile = voters_profile.rename(columns = {'index': 'province'})

# Returns the dataframe
voters_profile
```

```python
# Replaces the underscore line in the province name with white space

for i in range(len(voters_profile['province'])):
    white_space = voters_profile['province'][i].replace("_"," ")
    voters_profile['province'][i] = white_space
    
voters_profile['province']
```
