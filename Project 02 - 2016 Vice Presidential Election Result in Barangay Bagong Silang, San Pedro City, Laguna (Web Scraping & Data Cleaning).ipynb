```python
import pandas as pd
```

```python
url ='https://eleksyondata3.gmanews.tv/all_lvgs_results/REGION_IV-A_LAGUNA_SAN_PEDRO_CITY_BAGONG_SILANG.json.gz'
df = pd.read_json(url)
df
```

```python
vp_election_result = df['result'][1]
vp_election_result
```

```python
vp_election_result = vp_election_result.get('candidates')
vp_election_result
```

```python
vp_election_result = pd.DataFrame(vp_election_result)
vp_election_result.drop('party', axis='columns', inplace=True)
vp_election_result
```

```python
vp_election_result['party'] = vp_election_result['name'].str[-4:-1]
vp_election_result
```

```python
vp_election_result['party'] = vp_election_result['party'].str.strip('(')
vp_election_result
```

```python
vp_election_result['name'] = vp_election_result['name'].str.strip('(INDLPUA)')
vp_election_result
```

```python
second_column = vp_election_result.pop('party')
vp_election_result.insert(1,'party', second_column)
vp_election_result
```
